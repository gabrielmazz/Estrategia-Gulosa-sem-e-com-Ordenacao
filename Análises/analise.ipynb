5.2. Implementação

A implementação dos algoritmos se concentrou na criação de dois códigos interligados por um terceiro. O primeiro código, `estrategias.py`, é responsável por implementar as estratégias de resolução do problema da mochila da forma mais padrão possível. O segundo código, `estrategias_numba.py`, é responsável por implementar as mesmas estratégias, porém utilizando a biblioteca Numba para otimização de desempenho. Por fim, o código `main.py` é responsável por realizar a execução dos algoritmos, sendo possivel escolher qual será a forma de sua execução, se caso utilizara a biblioteca Numba ou não, ainda sendo possivel escolher se estarão ordenados ou não.

5.2.1. Como foi implementada o código `estrategias.py`

O código `estrategias.py` foi implementado de forma a conter duas funções principais, `estrategia_gulosa` e `estrategia_gulosa_ordenada`. Nas duas funções, suas implementações são basicamente as mesmas, começando calculando a razão entre o valor e o peso de cada item, colocando-os em uma array, nesta parte do algoritmo, vemos a unica diferença nas duas estratégias, onde na estratégia_gulosa_ordenada, a array é ordenada antes de ser utilizada, utilizando a função `sorted` do python. Depois deste processo, os dois algoritmos seguem o mesmo caminho, onde é feito um loop para adicionar os itens na mochila, sempre verificando se o peso do item é menor que o peso restante da mochila, e se for, adiciona o item na mochila e subtrai o peso do item do peso restante da mochila.

5.2.2. Como foi implementada o código `estrategias_numba.py`

O código `estrategias_numba.py` também foi implementado para conter as duas funções principais, `estrategia_gulosa` e `estrategia_gulosa_ordenada`. A diferença entre os dois arquivos, é a utilizaçao da biblioteca Numba, que é utilizada para otimizar o desempenho do código, fazendo com que o código seja executado de forma mais rápida. Ela traduz funções Python para código de máquina otimizado em tempo de execução, o que pode resultar em melhorias significativas de desempenho, otimizando loops e principalmente operações matemáticas, sendo um grande diferencial da implementação comum do problema da mochila.

5.2.3. Função `sorted` do Python

A função `sorted` do Python é uma função que retorna uma nova lista contendo todos os itens da sequência de entrada em ordem crescente ou decrescente. A função é uma mistura de algoritmos de ordenação, sendo ele o algoritmo Timsort, que é um algoritmo híbrido derivado do mergesort e do insertion sort, sendo um algoritmo de ordenação estável, o que significa que a ordem relativa dos itens iguais não é alterada. Com uma complexidade de tempo no pior caso de O(n log n), melhor caso de O(n) e caso médio de O(n log n), sendo um algoritmo muito eficiente para a ordenação de arrays.

5.2.3. Como é implementada a biblioteca Numba

Como dito anteriormente, a biblioteca Numba é uma biblioteca que otimiza o desempenho do código, fazendo com que o código seja executado de forma mais rápida. O Numba traduz funções Python para código de máquina otimizado em tempo de execução usando a biblioteca do compilador LLVM padrão da indústria. Algoritmos numéricos compilados pelo Numba em Python podem se aproximar das velocidades de C ou FORTRAN. Como o código é altamente otimizado em linguagem de máquina, aproveita-se o desempenho do hardware subjacente.

Tendo em vista os resultados obtidos, os algoritmos implementadados nas quatro situações possuem as mesma complexidade de tempo, e uma mesma função assintótica, chegando a $20n^2 + 44n + 5$, para os não ordenados e $20n^2 + 44n + 5$ + $nlogn$, para os ordenados, sendo que a diferença entre os dois é a adição da complexidade de tempo da função `sorted`. Em todos os caso a complexidade de tempo é quadrática ($O(n^2)$).
```python
# Estrutura correta dos dados para a tabela
table_data = {
    "Tamanho da Mochila": [10, 50, 100, 200, 300, 500, 750, 1000, 1250, 1500, 2000, 2500, 3000, 4000, 5000, 10000, 20000, 30000, 50000, 100000],
    "Guloso sem ordenação": Mochila,
    "Guloso com ordenação": Mochila_ordenacao,
    "Guloso com Numba e sem ordenação": Mochila_com_numba,
    "Guloso com Numba e com ordenação": Mochila_com_numba_ordenacao
}

# Crianção do DataFrame
df = pd.DataFrame(table_data)

# Transforma os dados para o formato longo
df_long = pd.melt(df, id_vars=["Tamanho da Mochila"], var_name="Algoritmo", value_name="Tempo (s)")
```

```python
# Ajusta o tamanho do iframe
display(HTML("<style>.container { width:100% !important; }</style>"))
display(HTML("<style>iframe { height: 1300px !important; }</style>"))

# ==========================================
# Criação do Dash
# ==========================================

# Inicialização do Dash
app = dash.Dash(__name__)

# Layout do Dash
app.layout = html.Div([
    html.H1("Comparação de Métodos de Mochila"),
    
    # Filtro interativo para selecionar os tamanho da mochila (eixo x)
    dcc.Checklist(
        id='x-filter', 
        options=[{'label': str(i), 'value': i} for i in df['Tamanho da Mochila']],
        value=df['Tamanho da Mochila'].tolist(),
        inline=True
    ),
    
    # Gráfico de linhas
    dcc.Graph(id='line-plot')
])

# Callback para atualizar o gráfico de linhas
@app.callback(
    Output('line-plot', 'figure'),
    [Input('x-filter', 'value')]
)
def update_line_plot(selected_values):
    # Filtra os dados
    filtered_df = df_long[df_long['Tamanho da Mochila'].isin(selected_values)]
    
    # Criação do gráfico de linhas com cores diferentes para cada linha
    traces = []
    for i, label in enumerate(filtered_df['Algoritmo'].unique()):
        filtered_data = filtered_df[filtered_df['Algoritmo'] == label]
        trace = go.Scatter(
            x=filtered_data['Tamanho da Mochila'],
            y=filtered_data['Tempo (s)'],
            mode='lines+markers+text',
            name=label,
            text=filtered_data['Tempo (s)'],
            textposition='top center',
            line=dict(color=px.colors.qualitative.Plotly[i])  # Cor da linha
        )
        traces.append(trace)
    
    # Criação da figura
    fig = px.line(
        filtered_df,
        x='Tamanho da Mochila',
        y='Tempo (s)',
        color='Algoritmo',
        title='Tempo de Execução para Diferentes Métodos',
        labels={'Tamanho da Mochila': 'Tamanho da Mochila', 'Tempo (s)': 'Tempo de execução (segundos)'},
    )
    
    # Atualização do layout
    fig.update_layout(
        title=dict(
            x=0.5,
            xanchor='center',
            yanchor='top',
            font=dict(size=28)),
        
        xaxis=dict(
            title='Tamanho da Mochila',
            titlefont=dict(size=24),
            tickfont=dict(size=18),),
        yaxis=dict(
            title='Tempo de execução (segundos)',
            showgrid=True,
            gridcolor='lightgray',
            titlefont=dict(size=24),
            tickfont=dict(size=18),),
        
        plot_bgcolor='white',
        paper_bgcolor='white',
        
        width=2000,
        height=1000
    )
    
    fig.update_traces(mode='lines+markers+text', marker=dict(size=10))
    
    return fig

# Execução do Dash
if __name__ == '__main__':
    app.run_server(debug=True)
```
